# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import numpy as np
import matplotlib.pyplot as plt


x_vals = np.linspace(-2, 3, 400)


y1_vals = x_vals * np.exp(-x_vals**2)
y2_vals = x_vals + 1


plt.plot(x_vals, y1_vals, label='y = xe^(-x^2)')
plt.plot(x_vals, y2_vals, label='y = x + 1')
plt.axvline(x=2, color='red', label='x = 2')


plt.fill_between(x_vals, y1_vals, y2_vals, where=(y1_vals < y2_vals) & (x_vals <= 2), color='lightblue')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico das Curvas')
plt.legend()


plt.show()
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração. 

##a)
$$\int_0^2 x^2+1 \space dx$$

##b) 
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



```python
#A)
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = x*2 + 1

a = 0
b = 2

f_np = sp.lambdify(x, f)

x_vals = np.linspace(a, b, 100)
y_vals = f_np(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.fill_between(x_vals, y_vals, where=((a <= x_vals) & (x_vals <= b)), alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')
plt.legend()
plt.show()

area = sp.integrate(f, (x, a, b))
print("valor da integral é:", area)
```

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x = sp.symbols('x')
f = 6*x**2 - 5*x + 2

a = -3
b = 1

f_np = sp.lambdify(x, f)

x_vals = np.linspace(a, b, 100)
y_vals = f_np(x_vals)

plt.plot(x_vals, y_vals, label='f(x)')
plt.fill_between(x_vals, y_vals, where=((a <= x_vals) & (x_vals <= b)), alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gráfico da função')
plt.legend()
plt.show()

area = sp.integrate(f, (x, a, b))
print("A área sobre a curva é:", area)
```

#Questão 5 (vale até 1 ponto)
Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou. 

Determinar a antiderivada da função f(x):
$ F(x) = \int f(x) \, dx = \int (3x^3 - 2x + 1) \, dx $

Resolver a integral indefinida:
$ F(x) = \frac{3}{4}x^4 - x^2 + x + C $\\

Substituir x por 1 na expressão de F(x) e igualar a zero:
$ F(1) = \frac{3}{4}(1)^4 - (1)^2 + 1 + C = 0 $

Resolver a equação para encontrar o valor de C:
$ \frac{3}{4} - 1 + 1 + C = 0 $
$ \frac{3}{4} + C = 0 $ \\
Resposta:
$ C = -\frac{3}{4} $
